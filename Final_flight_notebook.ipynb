```python
# cleaning the fat column
#checking for unique values in operators column
print(df['fat'].unique())
```

```python
#summing up errors with '5+1'in our data
df.loc[:,'fat'] = df['fat'].apply(lambda x: pd.eval(str(x)) if isinstance(x, str) else x)

#checking our data
print(df['fat'].unique())
```
